# arXiv EDA for ResearchRAG AI

This notebook performs exploratory data analysis on the processed arXiv subset used by the ResearchRAG AI project. It supports the fellowship requirements for data understanding, visualization, and documentation.

```python
import json
from collections import Counter
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

sns.set_theme(style='whitegrid')
DATA_PATH = Path('../data/processed_text/arxiv_documents.jsonl')
```

```python
records = []
with DATA_PATH.open('r', encoding='utf-8') as input_file:
    for line in input_file:
        if line.strip():
            records.append(json.loads(line))

df = pd.DataFrame(records)
df['primary_category'] = df['categories'].fillna('').str.split().str[0]
df['title_length'] = df['title'].fillna('').str.len()
df['abstract_length'] = df['abstract'].fillna('').str.len()
df.head()
```

```python
summary = {
    'documents': len(df),
    'unique_primary_categories': df['primary_category'].nunique(),
    'avg_title_length': round(df['title_length'].mean(), 2),
    'avg_abstract_length': round(df['abstract_length'].mean(), 2),
}
summary
```

```python
top_categories = df['primary_category'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_categories.values, y=top_categories.index, palette='crest')
plt.title('Top 10 Primary arXiv Categories in the Processed Subset')
plt.xlabel('Document Count')
plt.ylabel('Primary Category')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['title_length'], bins=30, kde=True, color='steelblue')
plt.title('Distribution of Title Lengths')
plt.xlabel('Title Length (characters)')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['abstract_length'], bins=40, kde=True, color='darkorange')
plt.title('Distribution of Abstract Lengths')
plt.xlabel('Abstract Length (characters)')
plt.tight_layout()
plt.show()
```

## Suggested Talking Points

- The processed subset is highly imbalanced, with `astro-ph` dominating the label distribution.
- Abstracts are long enough to support both retrieval and supervised classification.
- Macro metrics should be emphasized during evaluation because of the class imbalance.
- The top-category setup is a practical compromise that makes model training feasible while preserving real research domains.
